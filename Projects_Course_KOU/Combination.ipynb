# Combination

A **combination** is a mathematical technique that determines the number of possible arrangements in a collection of items where the order of the selection does not matter.
> In combinations, you can select the items in any order. (order of the selection does not matter)
* The arrangements *ab* and *ba* are equal in **combinations** (considered as one arrangement), while in **permutations**, the arrangements are different.

image.png

**n**: the total number of elements in a set.

**k**: the number of selected objects (the order of the objects is not important).


.

**Note that:**
> C(n,k) = C(n, n-k)

# Example:

A question paper has two parts **P** and**Q**, each containing 10 questions. If a student needs to choose 8 from part P and 4 from part Q, in how many ways can he do that?

## Solution

n = 10 quastions

k1 = 8 quastions from part P

k2 = 4 quastions from part Q

> Number of ways to choose 8 questions from part P = C(10,8) = C(10,2)

> Number of ways to choose 4 questions from part Q = C(10,4) = C(10,6)

Total number of ways:

## Python Code

```python
from scipy.special import comb

# n: the total number of elements in a set.
# k: the number of selected objects (the order of the objects is not important).

n  = int(input('Total number of quastions in each part: '))
k1 = int(input('Number of quastions should be selected from part P: '))
k2 = int (input('Number of quastions should be selected from part Q: '))

# Total number of ways:
T = comb(n, k1) * comb(n,k2)

print(f'\nTotal number of ways = {T}')
```

```python
from math import factorial

def combination(n, k):
  return factorial(n) / (factorial(n-k)*factorial(k))

# n: the total number of elements in a set.
# k: the number of selected objects (the order of the objects is not important).

n  = int(input('Total number of quastions in each part: '))
k1 = int(input('Number of quastions should be selected from part P: '))
k2 = int (input('Number of quastions should be selected from part Q: '))

# Total number of ways:
T = combination(n, k1) * combination(n,k2)

print(f'\nTotal number of ways = {T}')
```
