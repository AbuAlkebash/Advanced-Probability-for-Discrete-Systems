# Conditional Probability - Bayes Theorm

In probability theory,
> **conditional probability** is a measure of the probability of an event  occurring given that another event has (by assumption, presumption, assertion or evidence) occurred.

> In otherwords: given **A** is true, what is the probability that **B** is also true.  [**P(B|A)**]


## Bayes Theorm
> Bayes' formula is an important method for computing conditional probabilities using the multiplication rule.

Where:

> P(B|A)              : probability of B occurring, given A occurs.

> P(B ∩ B) = P(A ∩ B) :probability of both A and B occurring.

> P(A)                :probability of A occurring.

Dividing by **P(A)** removes the probability of anything not **A** from sample space. (i.e reduce the sample space)

# Example

Approximately 1% of women aged 40-50 have breast cancer. A woman with breast cancer has a 90% chance of a positive test from a mammogram, while a woman without breast cancer has a 10% chance of a false positive result. What is the probability a woman has breast cancer given that she just had a positive test?

## Solution

Quastion: probability that the woman has a cancer given that the test result is positive.

**P( Woman has a cancer | Positive test ) = ???**

*language of probability.*

let:
* A: the woman has a cancer
* B: the test is positive

Q:  **P(A|B) = ?**

* The probability that the women aged (40-50) has a breast cancer is 0.01
> P(A) = 0.01
* The probability that the woman has a cancer with positive test is 0.90
> P(B|A) = 0.90
* The probability that the woman without breast cancer with positive test is 0.10
> P(B|A') = 0.10

Use the **multiplication rule** to compute the following probabilities:

> P(A ∩ B)  = P(A) * P(B|A) = 0.01 * 0.90 = 0.009

> P(A' ∩ B) = P(A') * P(B|A') = 0.99  * 0.1 = 0.099

Now:
**P(B)  = P(A ∩ B) + P(A' ∩ B)**
      = 0.009 + 0.099 = 0.108

Finally, the **conditional probability**:

**P(A|B)**  = P(B|A)*P(A) / P(B)

        = 0.90 * 0.01 / 0.108
        = 0.083
        = 8.3%

# Python Code

```python
# Quastion: P( Woman has a cancer | Positive test ) = ?


def multiplication_rule(P_A, P_B_A, P_B_Ac):

  # probability of not_A  (probability of women without cancer)
  P_Ac = 1 - P_A

  # Multiplication rule:
  P_A_and_B  = P_A * P_B_A
  P_Ac_and_B = P_Ac * P_B_Ac

  # Compute the evidence (Probability of test to be positive):
  P_B = P_A_and_B + P_Ac_and_B

  return P_B



# A = the woman have a cancer
# B = the test is positive
# P(A|B) = ?

# probability of A    -prior probability
P_A   = float(input('The probability that the women aged (40-50) has a breast cancer is (%): '))/100

# probability of B given A    -likelihood
P_B_A = float(input('The probability that the woman has a cancer with positive test is (%): '))/100

# probability of B given not_A
P_B_Ac = float(input('The probability that the woman without breast cancer with positive test is (%): '))/100


# Compute the evidence (Probability of test to be positive):
P_B = multiplication_rule(P_A, P_B_A, P_B_Ac)

# solving for conditional probability:
P_A_B = (P_B_A * P_A / P_B) *100


print(f'\nP( Woman has a cancer | Positive test ) = {P_A_B:.2f} %')
```
